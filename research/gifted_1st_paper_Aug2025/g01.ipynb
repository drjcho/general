```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load your dataset
data = pd.read_csv('./normalized_data_with_stddev_revised01.csv')  # Replace with your file path

# Filter the relevant columns for K-means clustering
clustering_data = data[['AveAca', 'AveSocio']]

# Define number of clusters; use a larger number to ensure small clusters for gifted students
num_clusters = 18 #Define 1.5% of students as gifted
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(clustering_data)

# Calculate the size of each cluster
cluster_sizes = data['Cluster'].value_counts()

# Calculate the threshold for 1.3% to 1.9% of the total dataset
threshold_low = int(len(data) * 0.013)
threshold_high = int(len(data) * 0.019)

# Identify clusters that fall within this size range
gifted_clusters = cluster_sizes[(cluster_sizes >= threshold_low) & (cluster_sizes <= threshold_high)].index

# Mark students in the identified clusters as gifted
data['Gifted'] = data['Cluster'].apply(lambda x: 1 if x in gifted_clusters else 0)

# Generate a scatter plot to visualize the clusters and highlight gifted students
plt.figure(figsize=(10, 6))
plt.scatter(data['AveAca'], data['AveSocio'], c=data['Gifted'], cmap='coolwarm', marker='o', label='Gifted')
plt.title('K-means Clustering: Gifted Identification')
plt.xlabel('AveAca')
plt.ylabel('AveSocio')
plt.legend()
plt.show()

# Save the dataset with the gifted classification
data.to_csv('data_with_gifted_classification.csv', index=False)
print("Dataset with gifted classification saved to 'data_with_gifted_classification.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Load your dataset
data = pd.read_csv('./normalized_data_with_stddev_revised01.csv')  # Replace with your file path

# Filter the relevant columns for GMM clustering
clustering_data = data[['AveAca', 'AveSocio']]

# Apply Gaussian Mixture Model with 2 components (clusters)
gmm = GaussianMixture(n_components=2, random_state=42)
data['Cluster'] = gmm.fit_predict(clustering_data)

# Calculate the size of each cluster
cluster_sizes = data['Cluster'].value_counts()

# Identify the smaller cluster (which should approximately contain 1.5% of students)
gifted_cluster = cluster_sizes.idxmin()

# Mark students in the smaller cluster as gifted
data['Gifted'] = data['Cluster'].apply(lambda x: 1 if x == gifted_cluster else 0)

# Generate a scatter plot to visualize the clusters and highlight gifted students
plt.figure(figsize=(10, 6))
plt.scatter(data['AveAca'], data['AveSocio'], c=data['Gifted'], cmap='coolwarm', marker='o', label='Gifted')
plt.title('Gaussian Mixture Model (GMM): Gifted Identification (1.5% as Gifted)')
plt.xlabel('AveAca')
plt.ylabel('AveSocio')
plt.legend()
plt.show()

# Save the dataset with the gifted classification
data.to_csv('data_with_gifted_classification_gmm.csv', index=False)
print("Dataset with gifted classification saved to 'data_with_gifted_classification_gmm.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import numpy as np

# Load your dataset
data = pd.read_csv('./normalized_data_with_stddev_revised01.csv')  # Replace with your file path

# Filter the relevant columns for GMM clustering
clustering_data = data[['AveAca', 'AveSocio']]

# Apply Gaussian Mixture Model with 2 components (clusters)
gmm = GaussianMixture(n_components=2, random_state=42)
data['Cluster'] = gmm.fit_predict(clustering_data)

# Get the probabilities of each student belonging to the clusters
probabilities = gmm.predict_proba(clustering_data)

# We will assume the first cluster corresponds to non-gifted students and the second to gifted
# Sort by the probability of being in the second cluster
data['Gifted_Prob'] = probabilities[:, 1]

# Sort by the 'Gifted_Prob' to find the top 1.5% students
data = data.sort_values(by='Gifted_Prob', ascending=False)

# Calculate the number of students that should be identified as gifted (1.5%)
num_gifted = int(0.015 * len(data))

# Assign the top 1.5% as gifted
data['Gifted'] = 0
data.iloc[:num_gifted, data.columns.get_loc('Gifted')] = 1

# Generate a scatter plot to visualize the clusters and highlight gifted students
plt.figure(figsize=(10, 6))
plt.scatter(data['AveAca'], data['AveSocio'], c=data['Gifted'], cmap='coolwarm', marker='o', label='Gifted')
plt.title('Gaussian Mixture Model (GMM): Gifted Identification (Top 1.5% as Gifted)')
plt.xlabel('AveAca')
plt.ylabel('AveSocio')
plt.legend()
plt.show()

# Save the dataset with the gifted classification
data.to_csv('data_with_gifted_classification_gmm_1.5.csv', index=False)
print("Dataset with 1.5% gifted classification saved to 'data_with_gifted_classification_gmm_1.5.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import numpy as np

# Load your dataset
data = pd.read_csv('./normalized_data_with_stddev_revised01.csv')  # Replace with your file path

# Apply weights to AveAca and AveSocio
data['Weighted_Score'] = 0.7 * data['AveAca'] + 0.3 * data['AveSocio']

# Reshape the dataset to use the weighted score for clustering
clustering_data = data[['Weighted_Score']]

# Apply Gaussian Mixture Model with 2 components (clusters)
gmm = GaussianMixture(n_components=2, random_state=42)
data['Cluster'] = gmm.fit_predict(clustering_data)

# Get the probabilities of each student belonging to the clusters
probabilities = gmm.predict_proba(clustering_data)

# Assume the second cluster corresponds to gifted students
# Sort by the probability of being in the second cluster
data['Gifted_Prob'] = probabilities[:, 1]

# Sort by 'Gifted_Prob' to find the top 1.5% students
data = data.sort_values(by='Gifted_Prob', ascending=False)

# Calculate the number of students that should be identified as gifted (1.5%)
num_gifted = int(0.015 * len(data))

# Assign the top 1.5% as gifted
data['Gifted'] = 0
data.iloc[:num_gifted, data.columns.get_loc('Gifted')] = 1

# Generate a scatter plot to visualize the clusters and highlight gifted students
plt.figure(figsize=(10, 6))
plt.scatter(data['AveAca'], data['AveSocio'], c=data['Gifted'], cmap='coolwarm', marker='o', label='Gifted')
plt.title('GMM Clustering with Weights (Top 1.5% as Gifted)')
plt.xlabel('AveAca')
plt.ylabel('AveSocio')
plt.legend()
plt.show()

# Save the dataset with the gifted classification
data.to_csv('data_with_gifted_classification_gmm_weighted.csv', index=False)
print("Dataset with 1.5% gifted classification and weights saved to 'data_with_gifted_classification_gmm_weighted.csv'")
```
